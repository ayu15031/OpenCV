```python
import cv2
import numpy as np
import pandas as pd


face_classifier = cv2.CascadeClassifier('../data/haarcascade_frontalface_default.xml')

def face_detector(img, size=0.5):
    
    # Convert image to grayscale
    gray = cv2.cvtColor(img,0)
    faces = face_classifier.detectMultiScale(gray, 1.3, 5)
    if faces is ():
        return img, [], []
    
    roi = []
    loca = []
    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,255),2)
        temp = img[y:y+h, x:x+w]
        temp = cv2.resize(temp, (200, 200))
        loca.append((x,y))
        roi.append(temp)
    return img, roi, loca

model = cv2.face.LBPHFaceRecognizer_create()
model.read("faces/model/mymodel.yml")
# Open Webcam
cap = cv2.VideoCapture(0)

while True:

    ret, frame = cap.read()
    
    image, roi, loca = face_detector(frame)
#     print(len(roi))
    
    try:
        for i, face in enumerate(roi):
            face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)
            results = model.predict(face)
            df = pd.read_csv('faces/labels.csv')
        
            if results[1] < 500:
                confidence = int( 100 * (1 - (results[1])/400) )

            if confidence > 75:
                name_label = df.loc[df.id == results[0]].Name[results[0]-1]
                cv2.putText(image, name_label, loca[i], cv2.FONT_HERSHEY_COMPLEX, 1, (255,120,150), 2)
    #             cv2.putText(image, "Unlocked", (250, 450), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,0), 2)

            else:
                cv2.putText(image, "Unknown", loca[i], cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 2)

    except:
        cv2.putText(image, "No Face Found", (220, 120) , cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 2)
        pass

    cv2.imshow('Face Recognition', image)
    if cv2.waitKey(1) == ord('q'): 
        break
        
cap.release()
cv2.destroyAllWindows()     
```
