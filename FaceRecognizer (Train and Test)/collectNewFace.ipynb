```python
import os
import cv2
import numpy as np
import pandas as pd
import csv


# Load HAAR face classifier
face_classifier = cv2.CascadeClassifier('../data/haarcascade_frontalface_default.xml')
f = open("faces/labels.csv", 'a+')

if os.path.getsize("faces/labels.csv") == 0:
    writer = csv.DictWriter(
    f, fieldnames=["id", "Name"])
    writer.writeheader()
    id = 1
else:
    id = int(pd.read_csv("faces/labels.csv").iloc[-1][0])+1

string = input("Enter the name of person")

writer = csv.writer(f)
writer.writerow([id,string])
f.close()

def face_extractor(img):
    
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(gray, 1.3, 5)
    
    if faces is ():
        return None
    
    # Crop all faces found
    for (x,y,w,h) in faces:
        cropped_face = img[y:y+h, x:x+w]

    return cropped_face

# Initialize Webcam
cap = cv2.VideoCapture(0)
count = 0

while count<100:

    ret, frame = cap.read()
    if face_extractor(frame) is not None:
        count += 1
        face = cv2.resize(face_extractor(frame), (200, 200))
        face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)

        file_name_path = './faces/'+string+'.'+str(id)+'.'+str(count)+'.jpg'
        cv2.imwrite(file_name_path, face)

        cv2.putText(face, str(count), (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,0), 2)
        cv2.imshow('Face Cropper', face)
        
    else:
        print("Face not found")
        pass

    if cv2.waitKey(1) == ord('q') or count == 100: 
        break
        
cap.release()
cv2.destroyAllWindows()      
print("Collecting Samples Complete")
```
